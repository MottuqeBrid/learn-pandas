# Decision Tree

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeClassifier
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder,  OneHotEncoder
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report
```

```python
# load dataset
data = pd.read_csv("Churn_Modelling.csv")
data.head()
```

```python
data.shape
```

```python
data.isna().sum()
```

```python
data = data.drop(['RowNumber', 'CustomerId', 'Surname'], axis=1)
data.head()
```

```python
data.shape
```

```python
le_gender = LabelEncoder()
data["Gender"] = le_gender.fit_transform(data["Gender"]) # Male=1, Female=0
data.head()
```

```python
data.columns
```

```python
data["Geography"].value_counts()
```

```python
one_geography = OneHotEncoder(drop="first")
geo_encoded = one_geography.fit_transform(data[["Geography"]]).toarray()
geo_encoded_df = pd.DataFrame(geo_encoded, columns=one_geography.get_feature_names_out(["Geography"]))
geo_encoded_df
```

```python
data = pd.concat([data.drop("Geography", axis=1), geo_encoded_df], axis=1)
data.head()
```

```python
data.shape
```

```python
exited_count = data["Exited"].value_counts()
plt.figure(figsize=(6,4))
plt.pie(exited_count,
        labels=exited_count.index,
        autopct='%1.1f%%',
        startangle=90,
        colors=['lightblue', 'lightgreen'],
        explode=(0, 0.1),
        shadow=True)
plt.title("Distribution of Exited customers")
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

```python
data.corr()
```

```python
data.describe()
```

```python
x = data.drop("Exited", axis=1)
y = data["Exited"]
x.head()
```

```python
y.head()
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42)
x_train.shape, x_test.shape, y_train.shape, y_test.shape
```

```python
clf = DecisionTreeClassifier()
clf.fit(x_train, y_train)
```

```python
y_pred = clf.predict(x_test)
y_pred
```

```python
accuracy_score(y_test, y_pred)
```

```python
print(classification_report(y_test, y_pred))
```

```python
cm = confusion_matrix(y_test, y_pred)
cm
```

```python
cm_df = pd.DataFrame(
    cm,
    index=["Actual_Negative", "Actual_Positive"],
    columns=["Predicted_Negative", "Predicted_Positive"],
)
plt.figure(figsize=(6,4))
plt.imshow(cm_df, cmap='Blues')
plt.title("Confusion Matrix")
plt.xlabel("Predicted Label")
plt.ylabel("Actual Label")
plt.colorbar()
for (i, j), value in np.ndenumerate(cm_df.values):
    plt.text(j, i, value, ha='center', va='center', color='red')
```
