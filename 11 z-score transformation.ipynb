# Standarizatoin (Z Score)

```python
import pandas as pd
```

```python
data = {"Feature 1": [10, 20, 30, 40, 50], "Feature 2": [5, 15, 25, 35, 45]}
df = pd.DataFrame(data)
df
```

```python
def standardize_data(data):
    mean = data.mean()
    std = data.std(ddof=0) # Population standard deviation
    return (data - mean) / std
```

```python
standardize_df_raw = df.apply(standardize_data)
```

```python
standardize_df_raw
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
standardize_data = scaler.fit_transform(df)
pd.DataFrame(standardize_data, columns=df.columns)
```
