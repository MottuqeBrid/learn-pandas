# CountVectorizer

```python
import pandas as pd
import numpy as np
```

```python
dataset = [
    "I love my cat",
    "My cat is so cute",
    "My cat is so cute and lovely",
    "I love my dog",
    "My dog is so cute",
    "My dog is so cute and lovely"
]
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv = CountVectorizer()
x = cv.fit_transform(dataset)
```

```python
cv.get_feature_names_out()
```
