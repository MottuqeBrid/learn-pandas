```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
df=pd.read_csv("https://raw.githubusercontent.com/rashakil-ds/Linear-Regression-with-Python/refs/heads/main/nasdaq100.csv",sep=";")
df.head()
```

```python
df.isnull().sum()
```

```python
df=df.drop(columns=["Date"])
df.head()
```

```python
plt.scatter(df["Starting (USD)"],df["Ending (USD)"],color="red")
plt.xlabel("Starting Price")
plt.ylabel("Ending Price")
plt.title("Starting vs Ending Price")
plt.show()
```

```python
y=df[["Ending (USD)"]]
x=df[["Starting (USD)"]]
y.head()
```

```python
from sklearn.linear_model import LinearRegression
reg=LinearRegression()
```

```python
x.mean()
```

```python
y.mean()
```

```python
plt.scatter(x.mean(),y.mean(),color="green")
```

```python
reg.fit(x,y) #treaning the model
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
print(16700*reg.coef_+reg.intercept_)
reg.predict([[16700]])
```

```python
plt.plot(x,reg.predict(x),color="blue")
plt.scatter(x,y,color="red")
plt.plot(x.mean(),y.mean(),color="yellow",marker="o",markersize=10)
```
