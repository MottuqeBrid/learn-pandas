# Salary Prediction using Linear Regression

```python
import pandas as pd 
from matplotlib import pyplot as plt 
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/rashakil-ds/Linear-Regression-with-Python/refs/heads/main/Salary%20Data.csv", sep=',')
df.head()
```

# EDA

```python
df.isnull().sum()
```

```python
x=df[["Experience Years"]]
y=df[["Salary"]]
```

```python
plt.scatter(x,y)
plt.xlabel("Experience Years")
plt.ylabel("Salary")
```

# Linear Regression

```python
reg=LinearRegression()
reg.fit(x,y)
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
plt.plot(x,reg.predict(x),color='red')
plt.scatter(x,y)
plt.scatter(x.mean(),y.mean(),color='green',marker='o',s=100)
plt.ylabel("Salary")
plt.title("Experience Years vs Salary")
```

# Cost & Lose

```python
df["Lose"]=y-reg.predict(x)
df.head()
```

```python
reg.score(x,y)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error,r2_score
mean_absolute_error(y,reg.predict(x))
```

```python
mean_squared_error(y,reg.predict(x))
```

```python
r2_score(y,reg.predict(x))
```
