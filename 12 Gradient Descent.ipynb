```python
df = pd.read_csv("home data.csv")
df.head()
```

```python
df.shape
```

```python
x = df["x"]
y = df["y"]
```

```python
# first step
m = 0
c = 0
Lr = 0.001  # learning rate
n = int(len(x))  # number of data points
iterations = 100

# 2nd step
for i in range(iterations):
    y_pred = m * x + c  # the current predicted value of y
    deriv_m = (-2 / n) * sum(x * (y - y_pred))  # derivative wrt m\d
    deriv_c = (-2 / n) * sum(y - y_pred)  # derivative wrt c
    # final step - update m and c
    m = m - Lr * deriv_m
    c = c - Lr * deriv_c
    print(f"iteration {i}: m={m}, c={c}")
# print(f"final values: m={m}, c={c}")
```
