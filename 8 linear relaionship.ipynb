```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Generating linear data
x_linear = np.linspace(0, 10, 50)
y_linear=2*x_linear +5 +np.random.randn(50)*2 # adding random noise
```

```python
# Generating nonlinear data
x_nonlinear = np.linspace(0, 10, 50)
y_nonlinear = 2 * np.sin(x_nonlinear) + np.random.randn(50)   # adding random noise
```

```python
# linear
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.scatter(x_linear, y_linear, color='blue', label='Data with Linear relationship')
plt.plot(x_linear, 2*x_linear + 5, color='red', label='True Linear Regression')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Linear Relationship')
plt.legend()
plt.show()

# nonlinear
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.scatter(x_nonlinear, y_nonlinear, color='Green', label='Data with Nonlinear relationship')
plt.plot(x_nonlinear, 2 * np.sin(x_nonlinear), color='Orange', label='True Nonlinear Regression')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Nonlinear Relationship')
plt.legend()
plt.show()
```
